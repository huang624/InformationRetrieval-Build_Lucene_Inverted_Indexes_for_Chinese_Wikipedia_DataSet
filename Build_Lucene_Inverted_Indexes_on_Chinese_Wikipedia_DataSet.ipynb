# 資料匯入

```python
import json
# download dataset here https://drive.google.com/file/d/1yyqTwl3SCMQ6P6ZWt88cPqvtzK4RdI2m/view?usp=sharing
with open('./data/wiki_2021_08_05_1215639.json') as f:
  datas = json.load(f)
```

# 資料清洗

```python
for data in datas:
  data['id'] = str(data['id'])
  data['contents'] = data['articles']
  del data['title']
  del data['articles']
for i in range(10):
  print(datas[i])
```

# 將需要的資料存成json檔
# (建立反向索引時需要)

```python
with open(".resources/test.json", "w") as f:
  json.dump(datas, f)
```

# 建立反向索引表

```python
!python -m pyserini.index.lucene \
  --collection JsonCollection \
  --input .resources \
  --language zh \
  --index .indexes \
  --generator DefaultLuceneDocumentGenerator \
  --threads 1 \
  --storePositions --storeDocvectors --storeRaw
```

# 反向索引練習

```python
from pyserini.search.lucene import LuceneSearcher
import ast

searcher = LuceneSearcher('.indexes')
searcher.set_language('zh')
hits = searcher.search('烏克蘭')

for i in range(len(hits)):
    print(f'{i+1:2}. ')
    print(f'  score:{hits[i].score:.5f}')
    print(f'  id:{hits[i].docid}')
    doc = searcher.doc(hits[i].docid).raw()
    contents = ast.literal_eval(doc)["contents"]
    print(f'  contents:{contents}')
```

# 暴力搜尋法 V.S 反向索引法

```python
import time
search_word = '烏克蘭'

#Linear Search Time
start = time.time()
for data in datas:
  if search_word in data['contents']:
    continue
end = time.time()
print("Linear Search Time:", end - start)

#Inverted Index Search Time
start = time.time()
hits = searcher.search(search_word)
end = time.time()
print("Inverted Index Search Time:", end - start)
```
